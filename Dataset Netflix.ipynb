```python
#Apakah Netflix lebih fokus pada Acara TV daripada film dalam beberapa tahun terakhir?
```

```python
netflix.type.value_counts()
```

```python
sns.countplot(x= 'type', data=netflix)
plt.title('Count vs Type of Shows')
```

```python
#Memahami konten yang tersedia di berbagai negara
```

```python
netflix['country'].value_counts().head()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='country',order=netflix['country'].value_counts().index[0:10],data=netflix)
plt.title('Countrywise content on Netflix')
```

```python
#Menampilkan acara apa saja yang ditonton di semua negara 
```

```python
plt.figure(figsize=(12,4))
sns.countplot(y='listed_in',order=netflix['listed_in'].value_counts().index[0:5],data=netflix)
plt.title('Popular Genre on Netflix')
```

```python
#Mengurutkan Tahun Peluncuran dari yang terlama hingga terbaru 
```

```python
netflix['release_year'].value_counts()
```

```python
plt.figure(figsize=(40, 10))
sns.countplot(x= 'release_year', data=netflix)
plt.title('Release Year')
plt.xlabel('Year')
```
